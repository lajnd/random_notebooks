## Hierarchical Bayesian analysis for modeling BOLD time series data

```python
import matplotlib.pyplot as plt
import numpy as np
import pymc3 as pm 
import pandas as pd
import theano
import seaborn as sns
from sklearn.metrics import r2_score
from generatedata import genData
import scipy.stats as stats
```

```python
n_subs = 10
tr = 2
n_scans = 100
n_events = 24 
n_conds = 2
sub_idx = np.arange(n_subs)
```

Each participant will have their own beta estimate per condition, sampled from a common normal distribution

```python
mu = [3, 6] # cond 1 and 2
sigma = [1, 1] # cond 1 and 2

betas = []
for sub in range(n_subs):
    tmpbetas = []
    for cond in range(n_conds):
        tmpbetas.append(np.random.normal(mu[cond], sigma[cond], 1)[0])
    betas.append(tmpbetas)
    
data, X = genData(n_subs, tr, n_scans, betas, n_events)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(data['y'].values, alpha=0.2)
plt.plot(X['cond_0'].values, '--', color='g')
plt.legend(['Data', 'DM'], bbox_to_anchor=(1.1, 1.05))
plt.yticks(size=15)
plt.xticks(size=15)
sns.despine()
plt.show()
```

## Estimate betas by hierarchical bayesian MCMC

We define hyper parameters for the betas.
Hyper parameters for the means are sampled from a normal distribution around 0.
Since hyper parameters for SD cannot be negative, we sample those from a uniform distribution above 0.

```python
sub_idx = data['subject'].values
with pm.Model() as model:
    # hyper for b1
    hyper_b1_sd = pm.Uniform('hyper_b1_sd', lower=0, upper=5)
    hyper_b1_mu = pm.Normal('hyper_b1_mu', mu=0, sd=10)
    # hyper for b2
    hyper_b2_sd = pm.Uniform('hyper_b2_sd', lower=0, upper=5)
    hyper_b2_mu = pm.Normal('hyper_b2_mu', mu=0, sd=10)
    
    # Intercept
    b0 = pm.Normal('b0', mu=0, sd=10, shape=n_subs)
    # betas
    b1 = pm.Normal('b1', mu=hyper_b1_mu, sd=hyper_b1_sd, shape=n_subs)
    b2 = pm.Normal('b2', mu=hyper_b2_mu, sd=hyper_b2_sd, shape=n_subs)
    
    # Model error
    eps = pm.HalfCauchy('eps', beta=1)
    
    # Expected value
    y_hat = b0[sub_idx] + b1[sub_idx]*X['cond_0'] + b2[sub_idx]*X['cond_1']
    
    # Data likelihood
    y_like = pm.Normal('y_like', mu=y_hat[sub_idx], sd=eps, observed=data['y'])
    
    # Metropolis is quicker than NUTS, but yeah
    step = pm.Metropolis()
    trace = pm.sample(140000, step=step, progressbar=True)
```

# This is clearly not converging properly :(

```python
burnin = 10000
pm.traceplot(trace[burnin:], varnames=['b0','b1','b2'])
plt.show()
```

```python
fig, axes = plt.subplots(2, 5, figsize=(15,6))
axes = axes.ravel()
for sub in range(n_subs):
    
    ytmp = data[data.subject==sub]['y'].values
    Xtmp = X[X.subject==sub]

    b0_hat = trace['b0'][burnin:].mean()
    b1_hat = trace['b1'][burnin:, sub].mean()
    b2_hat = trace['b2'][burnin:, sub].mean()
    
    y_hat = b0_hat + b1_hat*Xtmp['cond_0'].values + b2_hat*Xtmp['cond_1'].values
    
    ax = axes[sub]
    ax.plot(ytmp, alpha=0.5)
    ax.plot(y_hat, '--', color='g')
    
    r2 = r2_score(ytmp, y_hat)
    ax.set_title(f'Subject {sub}\n$R^2$ = {r2:.4f}')
    sns.despine(ax=ax)
plt.tight_layout()
plt.show()
```

## Standard GLM fit

```python
fig, axes = plt.subplots(2, 5, figsize=(15,6))
axes = axes.ravel()
x = np.arange(n_scans)
for sub in range(n_subs):
    
    ytmp = data[data.subject==sub]['y'].values
    Xtmp = X[X.subject==sub]
    
    betastmp = np.linalg.lstsq(Xtmp[['cond_0', 'cond_1', 'constant']].values, ytmp, rcond=None)[0]
    y_hat = betastmp[-1] + betastmp[0]*Xtmp['cond_0'] + betastmp[1]*Xtmp['cond_1']
    
    ax = axes[sub]
    ax.plot(x, y_hat, '--', color='g')
    ax.plot(x, ytmp, alpha=0.5)
    
    
    r2 = r2_score(ytmp, y_hat)
    ax.set_title(f'Subject {sub}\n$R^2$ = {r2:.4f}')
    sns.despine(ax=ax)
plt.tight_layout()
plt.show()
```
