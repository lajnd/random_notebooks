## Standard and bayesian regression fit

```python
import matplotlib.pyplot as plt
import numpy as np
import pymc3 as pm 
import pandas as pd
import theano
import seaborn as sns
from nistats.design_matrix import make_design_matrix
from nistats.reporting import plot_design_matrix
from sklearn.metrics import r2_score
```

## Create design matrix

```python
# fmri settings
tr = 2
n_scans = 50
frame_times = np.arange(n_scans) * tr

# create design matrix
n_events = 24
events = pd.DataFrame()
events['duration'] = [1.0]*n_events
events['onset'] = np.sort(np.random.choice(frame_times[2:-2], n_events, replace=False))
events['trial_type'] = np.random.choice(['Cond_1', 'Cond_2'], 24)

X = make_design_matrix(
    frame_times, events, hrf_model='glover')

plt.figure(figsize=(8, 4))
plt.plot(X['Cond_1'], alpha=0.2)
plt.plot(X['Cond_2'], color='r', alpha=0.2)
plt.plot(X['Cond_1']+X['Cond_2'], '--', color='g')
plt.legend(['Cond 1', 'Cond 2', 'Combined'], bbox_to_anchor=(1.1, 1.05))
plt.yticks(size=15)
plt.xticks(size=15)
sns.despine()
plt.show()
```

## Add noise and create data

```python
# ground truth
true_b1 = 4
true_b2 = 2.3
# create data with noise
e = np.random.random(len(frame_times))*2
y = true_b1*X['Cond_1'] + true_b2*X['Cond_2'] + e

plt.figure(figsize=(8, 4))
plt.plot(y)
plt.plot(X['Cond_1']+X['Cond_2'], '--', color='g')
plt.legend(['Data (with noise)', 'Predicted'], bbox_to_anchor=(1.1, 1.05))
plt.yticks(size=15)
plt.xticks(size=15)
sns.despine()
plt.show()
```

## The convential way of finding your betas

```python
betas = np.linalg.lstsq(X, y, rcond=None)[0]
# print betas
print('Fitted values')
print(f'Intercept = {betas[-1]:.2f}, Beta 1 = {betas[0]:.2f}, Beta 2 = {betas[1]:.2f}')
print('True values')
print(f'Beta 1 = {true_b1}, Beta 2 = {true_b2}')

# plot
y_hat = betas[-1] + betas[0]*X['Cond_1'] + betas[1]*X['Cond_2']
plt.figure(figsize=(8, 4))
plt.plot(y)
plt.plot(y_hat, '--', color='g')
plt.legend(['Data', 'Predicted'], bbox_to_anchor=(1.1, 1.05))
plt.yticks(size=15)
plt.xticks(size=15)
r2 = r2_score(y, y_hat)
plt.title(f'Standard beta fit\n $R^2$ = {r2:.6f}')
sns.despine()
plt.show()
```

## Lets do the same thing but bayesian

We have 2 betas and one intercept to fit. We will draw all three values from their own normal distributions

```python
with pm.Model() as model:
    # Intercept
    b0 = pm.Normal('b0', mu=0, sd=20)
    # betas
    b1 = pm.Normal('b1', mu=0, sd=20)
    b2 = pm.Normal('b2', mu=0, sd=20)
    
    # Model error
    eps = pm.HalfCauchy('eps', beta=1)
    
    # Expected value
    y_hat = b0 + b1*X['Cond_1'] + b2*X['Cond_2']
    
    # Data likelihood
    y_like = pm.Normal('y_like', mu=y_hat, sd=eps, observed=y)
    
    # Use MAP to find a good starting poinnt
    start = pm.find_MAP()
    step = pm.Metropolis()
    trace = pm.sample(20000, step, start=start, progressbar=True)
```

```python
burnin = 2000
b0_hat = trace[burnin:].get_values('b0').mean()
b1_hat = trace[burnin:].get_values('b1').mean()
b2_hat = trace[burnin:].get_values('b2').mean()

pm.traceplot(trace[burnin:], varnames=['b0', 'b1', 'b2'], lines={'b0': b0_hat, 'b1': b1_hat, 'b2': b2_hat})
plt.show()
```

```python
# print betas
print('Fitted values')
print(f'Intercept = {b0_hat:.2f}, Beta 1 = {b1_hat:.2f}, Beta 2 = {b2_hat:.2f}')
print('True values')
print(f'Beta 1 = {true_b1}, Beta 2 = {true_b2}')

# plot
y_hat = b0_hat + b1_hat*X['Cond_1'] + b2_hat*X['Cond_2']
plt.figure(figsize=(8, 4))
plt.plot(y)
plt.plot(y_hat, '--', color='g')
plt.legend(['Data', 'Predicted'], bbox_to_anchor=(1.1, 1.05))
plt.yticks(size=15)
plt.xticks(size=15)
r2 = r2_score(y, y_hat)
plt.title(f'Bayesian beta fit\n $R^2$ = {r2:.6f}')
sns.despine()
plt.show()
```
